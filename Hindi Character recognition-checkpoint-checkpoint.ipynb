## Load data

```python
def load_data_from_directory(directory, class_names, description=None):
    #These lists will be populated with image data and corresponding labels during the function execution.
    all_images = []
    all_labels = []

#The function iterates over each class name in class_names using tqdm for progress visualization.
# For each class, it constructs the target directory path.
# images = os.listdir(target_dir): Retrieves the list of image files in the current class's directory.

    for class_name in tqdm(class_names,desc=description):
        target_dir = directory + "/" + class_name
        images = os.listdir(target_dir)# only names are here

# For each image file in the class:
# img_path = target_dir + "/" + img: Constructs the full path to the image.
# img_arr = cv2.imread(img_path, 0).flatten(): Reads the image using OpenCV, converts it to grayscale (0), and flattens it into a 1D NumPy array.
# img_label = class_names.index(class_name): Retrieves the label (class index) for the current image.

        for img in images:
            img_path = target_dir + "/" + img
            img_arr = cv2.imread(img_path,0).flatten()#first converting image into pixels then flatten it
            img_label = class_names.index(class_name)# notes index of that class from list of classes[class_names]

            all_images.append(img_arr.astype('float32')) #by defautlt images are stored in unsigned integer8; to enhance performance convert them into float32.
            all_labels.append(img_label) #Appends the image label to the list of all labels

    return np.array(all_images), np.array(all_labels) #Returns NumPy arrays containing all the processed images (all_images) and their corresponding 
                                                      #labels (all_labels).
```

```python
X_train, y_train = load_data_from_directory(train_dataset_path,class_names,"Loading train data...  ") # description for the progress bar
X_test, y_test = load_data_from_directory(test_dataset_path,class_names,"Loading test data...  ")
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Feature Selection : PCA

```python
from sklearn.decomposition import PCA 
# Import the PCA (Principal Component Analysis) class from scikit-learn, which is used for dimensionality reduction
```

```python
pca = PCA(n_components=0.98)  # Retain 98% of information PCA will retain principal components until they collectively explain 98% of the variance in the data.

# Fit and Transform Training Data:
X_train_pca = pca.fit_transform(X_train_scaled) #Fit the PCA model on the scaled training data (X_train_scaled) and then transform the data to its principal components.

# Transform Testing Data:
X_test_pca = pca.transform(X_test_scaled) #Transform the scaled testing data (X_test_scaled) using the same PCA instance fitted on the training data.
# This ensures consistency in dimensionality reduction.

# Reducing dimensionality using PCA can be beneficial in scenarios where the original feature space is high-dimensional, and it helps in speeding up the training of machine learning
# models while retaining most of the important information in the data.
```

```python
X_train_pca.shape, X_test_pca.shape
#shape of (m, k), it means there are m samples (data points) in the training set, and each sample is represented by k principal components.
```

```python
pca.n_components_ #no of principle components
```

# Model - 1 : Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Train Logistic Regression model(on training data)
logistic_model = LogisticRegression(max_iter=1000, n_jobs=-1)
#instance of the Logistic Regression model. Here are the parameters:
# max_iter=1000: Specifies the maximum number of iterations for the solver to converge.Reduce loss as much possible in 1000 trials
# n_jobs=-1: Utilizes all available processors for parallel computation during training for less time.

logistic_model.fit(X_train_pca, y_train)# giving image ans corresponding label to understand mapping by model . it learns pattern for future prediction.
# Fits (trains) the Logistic Regression model using the transformed training data X_train_pca obtained after PCA and the corresponding labels y_train.
```

```python
# Predict on the test set
y_pred = logistic_model.predict(X_test_pca)
#Uses the trained Logistic Regression model (logistic_model) to predict the target variable for the test data (X_test_pca). 
```

```python
accuracy_score(y_test,y_pred) 
# This line calculates the accuracy of the predictions made by the Logistic Regression model.
# y_test: The actual labels (ground truth) from the test set.
# y_pred: The predicted labels obtained from the Logistic Regression model.
```

## Hyper parameter tunning for Logistic regression 

```python
import optuna #library helps in hyperparameter tuning
```

```python
#main objective: find best value of main parameter C b/w  1e-5, 1e5 which will give best performance for my model
def objective(trial):
    # Define hyperparameters to search
    C = trial.suggest_float('C', 1e-5, 1e5, log=True)

    # Train Logistic Regression model with suggested hyperparameters
    logistic_model = LogisticRegression(C=C,max_iter=1000, n_jobs=-1)
    logistic_model.fit(X_train_pca, y_train)

    y_pred = logistic_model.predict(X_test_pca)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    return 1 - accuracy # returns error

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)# minimze the error in 50 trials

# Get the best hyperparameters
best_params = study.best_params
print("Best Hyperparameters:", best_params)

# Train the model with the best hyperparameters
best_model = LogisticRegression(C=best_params['C'], max_iter=1000, n_jobs=-1)
best_model.fit(X_train_pca, y_train)

# Evaluate the best model
y_pred_final = best_model.predict(X_test_pca)
final_accuracy = accuracy_score(y_test, y_pred_final)
print(f"Final Model Accuracy: {final_accuracy:.2f}")
```

### Hyper parameter tunning for KNN

```python
import optuna
from sklearn.model_selection import cross_val_score
```

```python
def objective(trial):
    # Define search space for hyperparameters
    n_neighbors = trial.suggest_int('n_neighbors', 1, 20)
    weights = trial.suggest_categorical('weights', ['uniform', 'distance'])
    
    # Initialize KNN classifier with hyperparameters
    knn_classifier = KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights, n_jobs=-1)

    # Use cross-validation for training and evaluation
    cv_score = cross_val_score(knn_classifier, X_train_pca, y_train, cv=5, scoring='accuracy').mean()

    return cv_score
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)
```

```python
# Get the best hyperparameters
best_params = study.best_params
print("Best Hyperparameters:", best_params)
```

```python
# Train KNN classifier with the best hyperparameters
best_knn_classifier = KNeighborsClassifier(n_neighbors=best_params['n_neighbors'] , weights=best_params['weights'],n_jobs=-1)
best_knn_classifier.fit(X_train_pca, y_train)
```

```python
# Evaluate the performance on the test set
y_pred = best_knn_classifier.predict(X_test_pca)
accuracy = accuracy_score(y_test, y_pred)
print(f"Final Model Accuracy: {accuracy:.2%}")
```

### Hyper parameter tunning for SVC

```python
import optuna
from sklearn.model_selection import cross_val_score
```

```python
def objective(trial):
    # Define search space for hyperparameters
    C = trial.suggest_float('C', 1e-5, 1e5,log=True)
    kernel = trial.suggest_categorical('kernel', ['linear', 'rbf', 'poly'])   # , 'sigmoid'
    degree = trial.suggest_int('degree', 2, 5) # if kernel == 'poly' else None
    gamma = trial.suggest_categorical('gamma', ['scale', 'auto']) # if kernel in ['rbf', 'poly', 'sigmoid'] else 'scale'

    # C = trial.suggest_float('C', 1e-3, 1e3, log=True)
    # kernel = trial.suggest_categorical('kernel', ['linear', 'rbf', 'poly'])
    # gamma = trial.suggest_float('gamma', 1e-3, 1e3, log=True)

    # Initialize SVC classifier with hyperparameters
    svc_classifier = SVC(C=C, kernel=kernel, degree=degree, gamma=gamma)

    # Use cross-validation for training and evaluation
    cv_score = cross_val_score(svc_classifier, X_train_pca, y_train, cv=5, scoring='accuracy').mean()

    return cv_score
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50, n_jobs=-1)
```

```python
# Get the best hyperparameters
best_params = study.best_params
print("Best Hyperparameters:", best_params)
```

```python
best_svc_classifier = SVC(C=best_params['C'],kernel=best_params['kernel'],degree=best_params['degree'],gamma=best_params['gamma'])
best_svc_classifier.fit(X_train_pca,y_train)
```

```python
# Evaluate the performance on the test set
y_pred = best_svc_classifier.predict(X_test_pca)
accuracy = accuracy_score(y_test, y_pred)
print(f"Final Model Accuracy: {accuracy:.2%}")
```
